```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

import plotly.io as pio
pio.renderers.default = "colab"
```

```python
x = np.linspace(0,5,11)
x
```

```python
y = x **3
y
```

```python
plt.plot(x,y, label="Line")

plt.title("The Title")

plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")

plt.legend()

plt.show()
```

```python
plt.plot(x,y, label="Line 1", color="blue", linestyle="--", marker="o")  # Line 1

plt.plot(x,y*2, label="Line 2", color="red", linestyle="-.", marker="s")  # Line 2

plt.plot(x,y*3, label="Line 3", color="green", linestyle="-", marker="D")  # Line 3

plt.title("Three Lines Example")

plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")

plt.legend()

plt.show()
```

```python
marker_options = [
    "o",  # Circle
    "s",  # Square
    "^",  # Triangle up
    "v",  # Triangle down
    "<",  # Triangle left
    ">",  # Triangle right
    "d",  # Diamond
    "D",  # Thick diamond
    "p",  # Pentagon
    "*",  # Star
    "h",  # Hexagon 1
    "H",  # Hexagon 2
    "+",  # Plus
    "x",  # X (cross)
    ".",  # Point
    ",",  # Pixel
]
```

```python
x1 = [1, 2, 3, 4, 5]
y_base = [2, 4, 6, 8, 10]

for i, m in enumerate(marker_options):
    plt.plot(x1, [y + i for y in y_base], label=f"Marker {m}", marker=m)

plt.title("Different Marker Options")
plt.legend()
plt.show()
```

```python
linestyle_options = [
    "-",    # Solid line
    "--",   # Dashed line
    "-.",   # Dash-dot line
    ":",    # Dotted line
]
```

```python
for i, ls in enumerate(linestyle_options):
    plt.plot(x1, [y + i for y in y_base], label=f"Linestyle {ls}", linestyle=ls)

plt.title("Different Line Styles")
plt.legend()
plt.show()
```

Sub Plots

```python
# 1.
plt.subplot(2,2,1)
plt.plot(x,y, label="Plot 1", color="blue")
plt.legend()
plt.title("Sub Plot 1")

# 2.
plt.subplot(2,2,2)
plt.plot(x**2,x**2, label="Plot 2", color="green")
plt.legend()
plt.title("Sub Plot 2")

# 3.
plt.subplot(2,2,3)
plt.plot(y,x, label="Plot 3", color="red")
plt.legend()
plt.title("Sub Plot 3")

# 4.
plt.subplot(2,2,4)
plt.plot(x,y, label="Plot 4", color="orange")
plt.legend()
plt.title("Sub Plot 4")

plt.tight_layout()
```

```python
fig = plt.figure()
# fig = plt.figure(figsize=(12,8), dpi=500) #including Figure Size and DPI

axis1 = fig.add_axes([0.1,0.1,1,1]) # [left, bottom, width, height]
axis1.plot(x,y, color="blue", label="Line 1")
axis1.set_title("Axis 1")
axis1.legend()

axis2 = fig.add_axes([0.3,0.7,.3,.3])
axis2.plot(x,y, color="red", label="Line 2")
axis2.set_title("Axis 2")
axis2.legend()

axis3 = fig.add_axes([0.3,0.3,.3,.3])
axis3.plot(x**2,x**2, color="green", label="Line 3")
axis3.set_title("Axis 3")
axis3.legend()
```

```python
x = np.linspace(0, 10, 10)
y = np.sin(x)
y2 = np.cos(x)
rand_data = np.random.randn(100)

# Generate random data for 3 groups for Violin plot
data = [
    np.random.normal(0, 1, 100),   # Group 1
    np.random.normal(5, 1.5, 100), # Group 2
    np.random.normal(10, 2, 100)   # Group 3
]
```

```python
plt.figure(figsize=(12, 10))

# 1. Line Plot
plt.subplot(3,3,1)
plt.plot(x, y, color="blue", label="sin(x)")
plt.plot(x, y2, color="red", label="cos(x)")
plt.title("Line Plot")
plt.legend()

# 2. Scatter Plot
plt.subplot(3,3,2)
plt.scatter(x, y, color="green", marker="o", label="sin(x)")
plt.scatter(x, y2, color="orange", marker="x", label="cos(x)")
plt.title("Scatter Plot")
plt.legend()

# 3. Bar Plot
plt.subplot(3,3,3)
plt.bar([1,2,3,4], [5,7,3,8], color="orange")
plt.title("Bar Plot")

# 4. Horizontal Bar Plot
plt.subplot(3,3,4)
plt.barh([1,2,3,4], [5,7,3,8], color="purple")
plt.title("Horizontal Bar Plot")

# 5. Histogram
plt.subplot(3,3,5)
plt.hist(rand_data, bins=20, color="skyblue", edgecolor="black")
plt.title("Histogram")

# 6. Boxplot
plt.subplot(3,3,6)
plt.boxplot(rand_data)
plt.title("Boxplot")

# 7. Stem Plot
plt.subplot(3,3,7)
plt.stem(x, y, linefmt="blue", markerfmt="ro", basefmt="k")
plt.title("Stem Plot")

# 8. Step Plot
plt.subplot(3,3,8)
plt.step(x, y, where="mid", color="brown")
plt.title("Step Plot")

plt.subplot(3,3,9)
plt.violinplot(data, showmeans=True, showmedians=True)
plt.title("Voilin Plot")

plt.tight_layout()
plt.show()
```
