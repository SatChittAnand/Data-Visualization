```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

import plotly.io as pio
pio.renderers.default = "colab"
```

```python
x = np.linspace(0,5,11)
x
```

```python
y = x **3
y
```

```python
plt.plot(x,y, label="Line")

plt.title("The Title")

plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")

plt.legend()

plt.show()
```

```python
plt.plot(x,y, label="Line 1", color="blue", linestyle="--", marker="o")  # Line 1

plt.plot(x,y*2, label="Line 2", color="red", linestyle="-.", marker="s")  # Line 2

plt.plot(x,y*3, label="Line 3", color="green", linestyle="-", marker="D")  # Line 3

plt.title("Three Lines Example")

plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")

plt.legend()

plt.show()
```

```python
marker_options = [
    "o",  # Circle
    "s",  # Square
    "^",  # Triangle up
    "v",  # Triangle down
    "<",  # Triangle left
    ">",  # Triangle right
    "d",  # Diamond
    "D",  # Thick diamond
    "p",  # Pentagon
    "*",  # Star
    "h",  # Hexagon 1
    "H",  # Hexagon 2
    "+",  # Plus
    "x",  # X (cross)
    ".",  # Point
    ",",  # Pixel
]
```

```python
x1 = [1, 2, 3, 4, 5]
y_base = [2, 4, 6, 8, 10]

for i, m in enumerate(marker_options):
    plt.plot(x1, [y + i for y in y_base], label=f"Marker {m}", marker=m)

plt.title("Different Marker Options")
plt.legend()
plt.show()
```

```python
linestyle_options = [
    "-",    # Solid line
    "--",   # Dashed line
    "-.",   # Dash-dot line
    ":",    # Dotted line
]
```

```python
for i, ls in enumerate(linestyle_options):
    plt.plot(x1, [y + i for y in y_base], label=f"Linestyle {ls}", linestyle=ls)

plt.title("Different Line Styles")
plt.legend()
plt.show()
```

Sub Plots

```python
# 1.
plt.subplot(2,2,1)
plt.plot(x,y, label="Plot 1", color="blue")
plt.legend()
plt.title("Sub Plot 1")

# 2.
plt.subplot(2,2,2)
plt.plot(x**2,x**2, label="Plot 2", color="green")
plt.legend()
plt.title("Sub Plot 2")

# 3.
plt.subplot(2,2,3)
plt.plot(y,x, label="Plot 3", color="red")
plt.legend()
plt.title("Sub Plot 3")

# 4.
plt.subplot(2,2,4)
plt.plot(x,y, label="Plot 4", color="orange")
plt.legend()
plt.title("Sub Plot 4")

plt.tight_layout()
```
